# ENTER DATA

Write the points in R^2, should put 1 as first coordinate (because we are working with quadrangles that aren't in the infinity)
Oordered starting at the bottom left and going counter-clock wise

```julia
p_a = [1 0 0]
p_b = [1 1 0]
p_c = [1 1 1]
p_d = [1 0 1]
```

# PROGRAM

```julia
function chessboard(p_a, p_b, p_c, p_d)
    #Find the intersections of the "parallel" sides of the quadrangle 
    if ( ((p_a[2] - p_b[2])*(p_d[3] - p_c[3])) == ((p_a[3] - p_b[3])*(p_d[2] - p_c[2])) )
        i_h = [0 (p_a[2] - p_b[2]) (p_a[3] - p_b[3])]

    else 
        M = [(p_b[2] - p_a[2]) (p_c[2] - p_d[2]); (p_b[3] - p_a[3]) (p_c[3] - p_d[3])]
        aux = [(p_d[2] - p_a[2]) (p_d[3] - p_a[3])]
        var = inv(M)*(aux')
        i_h = [1 (p_a[2] + var[1]*(p_b[2] - p_a[2])) (p_a[3] + var[1]*(p_b[3] - p_a[3]))]
    end
    if ( ((p_a[2] - p_d[2])*(p_b[3] - p_c[3])) == ((p_a[3] - p_d[3])*(p_b[2] - p_c[2])) )
        i_v = [0 (p_a[2] - p_d[2]) (p_a[3] - p_d[3])]

    else 
        M = [(p_d[2] - p_a[2]) (p_c[2] - p_b[2]); (p_d[3] - p_a[3]) (p_c[3] - p_b[3])]
        aux = [(p_b[2] - p_a[2]) (p_b[3] - p_a[3])]
        var = inv(M)*(aux')
        i_v = [1 (p_a[2] + var[1]*(p_d[2] - p_a[2])) (p_a[3] + var[1]*(p_d[3] - p_a[3]))]
    end

    #Calculate the multipliers to normalize the reference of the given quadrangle
    M = [p_a[1] i_h[1] i_v[1]; p_a[2] i_h[2] i_v[2]; p_a[3] i_h[3] i_v[3]]
    aux = p_c
    mult = inv(M)*(aux')

    #Matrix that goes from the square to the given quadrangle
    A = [(mult[1]*p_a[1]) (mult[2]*i_h[1]) (mult[3]*i_v[1]); (mult[1]*p_a[2]) (mult[2]*i_h[2]) (mult[3]*i_v[2]); (mult[1]*p_a[3]) (mult[2]*i_h[3]) (mult[3]*i_v[3])]


    #Output of the subdivision
    print(" SUBDIVISION: \n")

    point = [1 1/2 0]
    ps = point*(A')
    ps = ps/ps[1]
    print(point, " := ", ps, "\n")

    point = [1 1 1/2]
    ps = point*(A')
    ps = ps/ps[1]
    print(point, " := ", ps, "\n")

    point = [1 1/2 1]
    ps = point*(A')
    ps = ps/ps[1]
    print(point, " := ", ps, "\n")

    point = [1 0 1/2]
    ps = point*(A')
    ps = ps/ps[1]
    print(point, " := ", ps, "\n")

    point = [1 1/2 1/2]
    ps = point*(A')
    ps = ps/ps[1]
    print(point, " := ", ps, "\n")

    #The points of the chessboard given by columns
    print("\n \n CHESSBOARD \n \n")
    for i in 0:8 
        # Inner For-loop
        print("Column ", i, "\n")
        for j in 0:8 
            # Print statement
            pij = [1 (i/8) ((8-j)/8)]
            res = pij*(A')
            res = res/res[1]
            print(res)
            print(" \n")
        end
        println()
    end
end
```

```julia
chessboard(p_a, p_b, p_c, p_d)
```
